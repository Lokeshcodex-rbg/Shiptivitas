```python
!pip install yfinance bs4 plotly pandas lxml

import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go


def get_stock_data(ticker):
    data = yf.Ticker(ticker)
    df = data.history(period="max")
    df.reset_index(inplace=True)
    return df


def get_revenue_data(url):
    html_data = requests.get(url).text
    soup = BeautifulSoup(html_data, "html.parser")

    tables = soup.find_all("table")
    df = pd.read_html(str(tables[1]))[0]

    df.columns = ["Date", "Revenue"]
    df["Revenue"] = df["Revenue"].str.replace(",", "").str.replace("$", "")
    df.dropna(inplace=True)

    return df

# TESLA DATA

tesla_data = get_stock_data("TSLA")

tesla_revenue_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
tesla_revenue = get_revenue_data(tesla_revenue_url)

print("Tesla Stock Data:")
print(tesla_data.head())

print("\nTesla Revenue Data:")
print(tesla_revenue.tail())


# AMAZON DATA
amazon_data = get_stock_data("AMZN")
print("\nAmazon Stock Data:")
print(amazon_data.head())

# AMD DATA
amd_data = get_stock_data("AMD")
print("\nAMD Stock Data:")
print(amd_data.head())

# GAMESTOP DATA
gme_data = get_stock_data("GME")
print("\nGameStop Stock Data:")
print(gme_data.head())

# PLOT
def plot_graph(data, title):
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=data['Date'],
        y=data['Close'],
        name="Close Price"
    ))

    fig.update_layout(
        title=title,
        xaxis_title="Date",
        yaxis_title="Price"
    )

    fig.show()

plot_graph(tesla_data, "Tesla Stock Price")
plot_graph(amazon_data, "Amazon Stock Price")
plot_graph(amd_data, "AMD Stock Price")
plot_graph(gme_data, "GameStop Stock Price")

print("\nINSIGHTS:")
print("- Tesla shows strong growth in both revenue and stock price.")
print("- Amazon demonstrates steady long-term growth.")
print("- AMD shows gradual increase with some volatility.")
print("- GameStop shows extreme volatility around 2021.")
```
